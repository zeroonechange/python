```python
import numpy as np
import pandas as pd
data = pd.Series(np.random.randn(9),
                 index=[['a', 'a', 'a', 'b', 'b', 'c', 'c', 'd', 'd'],
                [1, 2, 3, 1, 3, 1, 2, 2, 3]])
data
```

```python
data.index     # 俩层索引  奇怪的很
```

```python
data['b']
```

```python
data.loc[:,2]
```

```python
data.unstack()  
```

```python
data.unstack().stack()  # 逆运算
```

```python
frame = pd.DataFrame(np.arange(12).reshape((4,3)), 
                    index=[['a', 'a', 'b', 'b'], [1,2,1,2]],
                    columns=[['Ohio', 'Ohio', 'Colorado'], ['Green', 'Red', 'Green']])
frame
```

```python
frame.index.names=['key1', 'key2']   # 给索引命名
frame.columns.names=['state', 'color']
frame
```

```python
frame['Ohio']   # 取列 
```

```python
frame.swaplevel('key1', 'key2')   # 调换索引的顺序
```

```python
frame.sort_index(level=1)  # 排序  这样会以key2来排序  
```

```python
frame.sum(level='key2')  # 对某个索引求和 
```

```python
frame.sum(level='color', axis=1)  # 对列来求和
```

```python
frame = pd.DataFrame({'a': range(7), 'b': range(7, 0, -1),
                      'c': ['one', 'one', 'one', 'two', 'two','two', 'two'],
                      'd': [0, 1, 2, 0, 1, 2, 3]})
frame
```

```python
frame2=frame.set_index(['c', 'd'], drop=True)   # 会将列设置为索引    drop是否丢弃原来的列
frame2
```

```python
frame2.reset_index()   # 逆操作
```
