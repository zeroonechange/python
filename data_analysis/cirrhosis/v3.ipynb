```python
from sympy import *
import numpy as np
import pandas as pd
data = pd.read_excel('RS_R.xls')
cleandata = data.iloc[:, [33,35]].dropna()
cleandata.tail()
```

```python
def unsolve(y):
    R = symbols('R')
    return solve(Eq(log(R)/(1-R), y),R)
cleandata['R1'] = cleandata['A111'].map(unsolve)
cleandata['R2'] = cleandata['A22'].map(unsolve)
cleandata
```

```python
cleandata.to_excel('result.xlsx')
```
