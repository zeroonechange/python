```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'],
                   'key2' : ['one', 'two', 'one', 'two', 'one'],
                   'data1' : np.random.randn(5),
                   'data2' : np.random.randn(5)})
df
```

```python
# 按key1进行分组，并计算data1列的平均值
grouped = df['data1'].groupby(df['key1'])
grouped   #是个GroupBy对象  还没有进行任何计算，只是含有一些有关分组键df['key1']的中间数据而已
```

```python
grouped.mean()   # 数据（Series）根据分组键进行了聚合，产生了一个新的Series，其索引为key1列中的唯一值。
```

```python
means = df['data1'].groupby([df['key1'], df['key2']]).mean()  # 取俩个作为索引
means
```

```python
# 使用自己的聚合函数    开销非常大  一般别用  体现在构造中间分组数据时函数调用和数据重排
def peak_to_ocean(arr):
    return arr.max() - arr.min()
df.groupby('key1').agg(peak_to_ocean)  # 使用自己的函数 求最大值-最小值
```

```python
# apply 的讲解
def top(df, n=5, column='tip_pct'):
    return df.sort_values(by=column)[-n:]
top(tips, n=6)   # 按照某列进行排序， 取最后n个  
```

```python
tips.groupby('smoker').apply(top)  # 先根据分组为NO YES    然后再进行选取操作
```

```python
#  原理就是top 函数在各个片段上被调用  然后由pandas.concat组装到一起，分组标记
tips.groupby(['smoker', 'day']).apply(top, n=1, column='total_bill')   
# 先分为俩个列索引  然后取组名称是total_bill 然后 top1 的值
```

```python
# 分位数和桶分析
frame = pd.DataFrame({'data1': np.random.randn(1000),
                                       'data2': np.random.randn(1000)})
frame.head()
```

```python
quartiles = pd.cut(frame.data1, 4)   # cut桶  得到大小相等的桶，使用qcut
quartiles[:10]  
```

```python
def get_stats(group):
    return {'min': group.min(), 'max': group.max(), 'count': group.count(), 'mean' : group.mean()}

grouped = frame.data2.groupby(quartiles)
grouped.apply(get_stats).unstack()
```

```python
grouping = pd.qcut(frame.data1, 10, labels=False)  # 使用qcut来分桶
grouped = frame.data2.groupby(grouping)
grouped.apply(get_stats).unstack()
```

```python
#  随机采样和排列
suits = ['H', 'S', 'C', 'D']
card_val = (list(range(1, 11)) + [10]*3) * 4
base_names = ['A'] + list(range(2, 11)) + ['J', 'K', 'Q']
cars = []
for suit in ['H', 'S', 'C', 'D']:
    cars.extend(str(num) + suit for num in base_names)
    
deck = pd.Series(card_val, index=cars)
deck[:13]
```

```python
def draw(deck, n=5):
    return deck.sample(n)

draw(deck)   # 随机取5个
```

```python
# 假设从每种花色各取俩张牌
get_suit = lambda card: card[-1]
deck.groupby(get_suit).apply(draw, n=2)
```

```python
# 分组加权平均数和相关系数    列与列之间的运算
df = pd.DataFrame({'category': ['a', 'a', 'a', 'a','b', 'b', 'b', 'b'],
                   'data': np.random.randn(8),
                   'weights': np.random.rand(8)})
df
```

```python
close_px = pd.read_csv('examples/stock_px_2.csv', parse_dates=True, index_col=0)
close_px.info()
```

```python
# 任务:  计算一个由日收益率（通过百分数变化计算）与SPX之间的年度相关系数组成的DataFrame
rets = close_px.pct_change().dropna()
spx_corr = lambda x: x.corrwith(x['SPX'])  # 虽然暂时不懂  后面总会懂的  只是个函数而已 
get_year = lambda y : y.year
by_year = rets.groupby(get_year)
by_year.apply(spx_corr)
```

```python
by_year.apply(lambda g: g['AAPL'].corr(g['MSFT']))  # 相关系数
```

```python
#  组级别的线性回归
# regress函数（利用statsmodels计量经济学库）对各数据块执行普通最小二乘法（Ordinary Least Squares，OLS）回归
import statsmodels.api as sm
def aggress(data, yvar, xvars):
    Y = data[yvar]
    X = data[xvars]
    X['intercept'] = 1.
    result = sm.OLS(Y, X).fit()
    return result.params
# 按年计算AAPL对SPX收益率的线性回归
by_year.apply(aggress, 'AAPL', ['SPX'])
```
